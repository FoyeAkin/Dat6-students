```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from ggplot import *
%matplotlib inline

d = pd.read_csv('/Users/heatherhardway/Desktop/DAT6/data/UNdata.csv')
np.random.seed(0)

print ggplot(d, aes(x='lifeMale', y='lifeFemale',colour='region'))+geom_point()+geom_text(aes(label='country'),hjust=0, vjust=0)
print ggplot(d, aes(x='lifeMale', y='infantMortality',colour='region'))+geom_point() +geom_text(aes(label='country'),hjust=0, vjust=0)
```

```python
print ggplot(d, aes(x='infantMortality', y='GDPperCapita',colour='region', label='country'))+geom_point()+geom_text()+ylim(0,45000)
print ggplot(d, aes(x='lifeMale', y='GDPperCapita',colour='region', label='country'))+geom_point()+geom_text()+ylim(0,45000)

d.head()
```

```python
est = KMeans(n_clusters=3, init='random')
y=est.fit(d.iloc[:,3:6])
```

```python
est.cluster_centers_
```

```python
d['predict'] = est.predict(d.iloc[:,3:6])
d.head()
```
